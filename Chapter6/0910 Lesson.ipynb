## 한국일보 RSS 사회면 기사를 수집하여 명사만 추출하기

* <a href= 'http://rss.hankooki.com/daily/dh_society.xml'>한국일보 RSS 사회면</a>(http://rss.hankooki.com/daily/dh_society.xml)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://rss.hankooki.com/daily/dh_society.xml'
content = requests.get(url).content
soup = BeautifulSoup(content, "xml")
urls = soup.select("item > link")
urls[0].text
```

```python
from konlpy.tag import Komoran
komoran = Komoran()
news = list()
for url in urls:
    new_url = url.text
    new_content = requests.get(new_url)
    new_soup = BeautifulSoup(new_content.content, "html.parser")
    gisa = new_soup.select_one('#GS_Content').text
    gisa = gisa.strip()
    gisa = gisa.replace('[', '')
    gisa = gisa.replace(']', '')
    gisa = gisa.replace('\n', '')
    gisa = gisa.replace('\t', '')
    gisa = gisa.replace('기자', '')
    news.append(gisa)
```

```python
string = ''
for newsone in news:
    string += newsone
```

```python
result = [word for word in komoran.nouns(string) if len(word) > 1]
result = set(result)
result
```
