```python
import pandas as pd
import re
```

```python
CCTV_Seoul = pd.read_excel('서울시 자치구 년도별 CCTV 설치 현황.xlsx')
```

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul['기관명'] = CCTV_Seoul['기관명'].apply(lambda x : re.sub(r'\s+', '', x))    #기관명값에 공백없애기
```

```python
CCTV_Seoul.rename(columns = {CCTV_Seoul.columns[0] : '구별'}, inplace = True)  #기관명을 구별로 수정
```

```python
CCTV_Seoul.sort_values(by = '소계', ascending = True).head(5)    # 소계별 오름차순 소팅
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2017년'] + CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + \
                             CCTV_Seoul['2014년']) / CCTV_Seoul['2013년 이전'] * 100
CCTV_Seoul.sort_values(by = '최근증가율', ascending = False).head(5)
```

```python
pop_Seoul = pd.read_excel('Report.xls', header = 2, parse_cols = 'B, D, G, J, N')
```

```python
pop_Seoul.head()
```

```python
pop_Seoul.columns = ['구별', '인구수', '한국인', '외국인', '고령자']    #컬럼 이름 변경
```

```python
pop_Seoul.head()
```

```python
pop_Seoul.drop([0], inplace = True)    #0번째 행 삭제
```

```python
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
pop_Seoul[pop_Seoul['구별'].isnull()]   #구별의 값에 null이 있는지 여부 파악
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by = '인구수', ascending = False).head()    #인구수 내림차순
```

```python
pop_Seoul.sort_values(by = '외국인', ascending = False).head()    #외국인 내림차순
```

```python
pop_Seoul.sort_values(by = '외국인비율', ascending = False).head()    #외국인비율 내림차순
```

```python
pop_Seoul.sort_values(by = '고령자', ascending = False).head()    #고령자 내림차순
```

```python
pop_Seoul.sort_values(by = '고령자비율', ascending = False).head()    #고령자비율 내림차순
```

```python
data = pd.merge(CCTV_Seoul, pop_Seoul, on = '구별')
data.head()
```

```python
del data['2013년 이전']
del data['2014년']
del data['2015년']
del data['2016년']
del data['2017년']
data.head()
```

```python
data.set_index('구별', inplace = True)
data.head()
```

```python
import numpy as np
```

```python
np.corrcoef(data['고령자비율'], data['소계'])   #고령자비율과 소계와의 상관계수 구하기
```

```python
np.corrcoef(data['외국인비율'], data['소계'])   #외국인비율과 소계와의 상관계수 구하기
```

```python
np.corrcoef(data['인구수'], data['소계'])   #인구수와 소계와의 상관계수 구하기
```
